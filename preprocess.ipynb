```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("contoh_csv_1.csv")
df
```

```python
index = df.columns[1:]
df = df.loc[:, index]
df
```

```python
dataset = df.to_numpy()
dataset.shape
```

```python
# Remove all minus data
for idx, dim in enumerate(dataset):
    for idx2, number in enumerate(dim):
        if number < 0:
            dataset[idx][idx2] = number * -1
```

```python
dataset = dataset.transpose()
dataset.shape
```

```python
import matplotlib.pyplot as plt

# index 4 : Jumalh kumulatif sembuh
# Index 2 : Jumlah dirawat perhari
plt.plot(dataset[4], dataset[2])
```
